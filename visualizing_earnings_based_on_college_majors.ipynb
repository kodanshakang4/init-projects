### Histogram 

```python
# To explore the distribution of column, we can select the column series from the DataFrame
recent_grads['Sample_size'].plot(kind='hist') 
# since plot() cannot be controlled bin size, hist() can be controlled
# bin length will be 5000/25 = 200
recent_grads['Sample_size'].hist(bins=25, range=(0,5000))
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]
# from Sample_size to Full_time
fig = plt.figure(figsize=(5,12))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40)
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]
# from ShareWomen to Women
fig = plt.figure(figsize=(5,12))
for r in range(4,8):
    ax = fig.add_subplot(4,1,r-3)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40)
    
```
